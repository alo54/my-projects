```python
from transformers import pipeline

# Load the sentiment analysis pipeline with BETO fine-tuned
classifier = pipeline("sentiment-analysis", model="nlptown/bert-base-multilingual-uncased-sentiment")

# Example COVID-19 headlines
headlines = [
    "Aumentan los casos de COVID-19 en América Latina",
    "La OMS celebra los avances en la vacuna contra el coronavirus",
    "Los hospitales están colapsando por la nueva ola de contagios",
    "Mejora la situación sanitaria en Europa tras meses de pandemia", 
    "COVID"
]

# Analyze and print results
results = classifier(headlines)

for h, r in zip(headlines, results):
    print(f"{h}\n→ {r['label']} ({r['score']:.2f})\n")
```
