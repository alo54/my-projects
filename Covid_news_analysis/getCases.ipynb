```python
import pandas as pd
import numpy as np


df = pd.read_csv("Casos_Diarios_Estado_Nacional_Confirmados_20230625.csv")
df
```

```python
df_window = df.loc[:, '15-03-2020':'30-04-2020']
df_window["total_cases"] = df_window.sum(axis = 1)
df_window['nombres'] = df["nombre"]
df_window["poblacion"] = df["poblacion"]
df_window["percentaje"] = df_window["total_cases"] * 100 / df_window["poblacion"]
df_window.set_index("nombres", inplace = True)
df_window
```
